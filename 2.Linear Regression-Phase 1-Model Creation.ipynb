```python
# Pandas is a library which contains functions related to database
import pandas as pd
```

```python
# Storing the required csv file dataset in the variable "dataset" 
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Spliting Input dataset from overall dataset in a variable "independent"
independent=dataset[["YearsExperience"]]
independent
```

```python
# Spliting Output dataset from overall dataset in a variable "independent"
dependent=dataset[["Salary"]]
dependent
```

```python
# sklearn is a library which contains functions related to machine learning

from sklearn.model_selection import train_test_split

# Splitting train & test datasets

X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
#Model_creation
#sklearn.linear_model is library which contains functions related to linear regression
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# weight is a variable which contains the slope of our model
weight=regressor.coef_
```

```python
weight
```

```python
# bais is a variable which contains the intercept or initial or starting point value of our model
bais=regressor.intercept_
```

```python
bais
```

```python
# predicitng the test datasets
Y_pred=regressor.predict(X_test)
```

```python
Y_pred
```

```python
#Evaluation_Metrics
# r2 score is a evaluation metrics fucntion which cross check the trained model's performance(good or bad)

from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
# If r score value is near to 1 then it is a good model. if it is near to 0 then it is a bad model

r_score
```

```python
#Save_the_Best_Model
# Pickle is a library which is used to save and load the model

import pickle
filename="Finalized_model_linear.sav"
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
